## OCR to convert images to text

Example questions from our dataset after processing through an OCR agent (Gemini 2.0 Flash)

### Example 1: Math question

```python
from PIL import Image
from google import genai

client = genai.Client(api_key='API_KEY')

prompt = ''' 
You are an expert OCR system specialized in mathematical content. Your task is to extract all text and mathematical expressions from the given image with high accuracy.

Instructions:

Transcribe ALL visible text exactly as it appears
For mathematical expressions, use proper LaTeX notation where appropriate
Preserve the structure and layout (line breaks, indentation, spacing)
If text is unclear or ambiguous, indicate with [UNCLEAR] but make your best attempt
Include all symbols, superscripts, subscripts, fractions, integrals, limits, etc.
Maintain the logical flow and organization of the content
Please transcribe the mathematical content from this image completely and accurately.

Pay special attention to:

Limit expressions and their notation
Differential equations and their solutions
Integration symbols and bounds
Fraction notation and complex expressions
Solution steps and logical flow indicators (arrows, "therefore" symbols)
'''

image = Image.open(r"question_images\2024\Paper_2_Hindi\Mathematics\2024_P2_Hindi_Mathematics_MCQ-Multiple_q1_MCQ-Multiple_page3.png")
response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=[image, prompt]
)
print(response.text)
```
